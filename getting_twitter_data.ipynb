```python
import tweepy
import pandas as pd
import json
from nltk.tokenize import word_tokenize
from config import api_key, secret_api_key, access_token, secret_access_token
```

```python
auth = tweepy.OAuthHandler(api_key, secret_api_key)
auth.set_access_token(access_token, secret_access_token)
api = tweepy.API(auth)
```

```python
# API.search(q, [geocode], [lang], [locale], [result_type], \
#                        [count], [until], [since_id], [max_id], \
#                        [include_entities])
response = api.search(q="Cheesecake Factory",count=100,tweet_mode="extended",lang="en")
```

```python
df_tweet_atomic = pd.DataFrame({
    "tweet_id":[i.id_str for i in response],
    "time_when_tweet_created":[i.created_at for i in response],
    "tweet_range":[i.display_text_range for i in response],
    "full_text":[i.full_text for i in response],
    "tweet_hastags":[i.entities["hashtags"] if i.place != None else None for i in response],
    "tweet_source":[i.source for i in response],
    "truncated":[i.truncated for i in response],
    "is_quoted_tweet_status":[i.is_quote_status for i in response],
    "retweet_count":[i.retweet_count for i in response],
    "favorite_count":[i.favorite_count for i in response],
    "lang":[i.lang for i in response]
})

# user object
df_tweet_user = pd.DataFrame({
    "tweet_id":[i.id_str for i in response],
    "user_id":[i.user.id for i in response],
    "name":[i.user.name for i in response],
    "screen_name":[i.user.screen_name for i in response],
    "location_name":[i.user.location for i in response],
    "url":[i.user.url for i in response],
    "description":[i.user.description for i in response],
    "verified":[i.user.verified for i in response],
    "followers_count":[i.user.followers_count for i in response],
    "friends_count":[i.user.friends_count for i in response],
    "listed_count":[i.user.listed_count for i in response],
    "favourites_count":[i.user.favourites_count for i in response],
    "statuses_count":[i.user.statuses_count for i in response],
    "time_when_account_created":[i.user.created_at for i in response],
    "time_zone":[i.user.time_zone for i in response],
    "geo_enabled":[i.user.geo_enabled for i in response],
    "language":[i.user.lang for i in response],
    "location_type_and_coordinates":[i.coordinates if i.place != None else None for i in response]
})

# place object
df_tweet_place = pd.DataFrame({
    "tweet_id":[i.id_str for i in response],
    "place_id":[i.place.id if i.place != None else None for i in response],
    "name":[i.place.full_name if i.place != None else None for i in response],
    "full_name":[i.place.full_name if i.place != None else None for i in response],
    "country":[i.place.country if i.place != None else None for i in response],
    "country_code":[i.place.country_code if i.place != None else None for i in response],
    "bounding_box":[i.place.bounding_box if i.place != None else None for i in response],
    "type":[i.place.place_type if i.place != None else None for i in response],
    "url":[i.place.url if i.place != None else None for i in response],
})
```

```python
# separate lats and longs in tweet_place
df_tweet_place["location_type"] = 0
df_tweet_place["long_1"] = 0
df_tweet_place["lat_1"] = 0
df_tweet_place["long_2"] = 0
df_tweet_place["lat_2"] = 0
df_tweet_place["long_3"] = 0
df_tweet_place["lat_3"] = 0
df_tweet_place["long_4"] = 0
df_tweet_place["lat_4"] = 0
for index, row in df_tweet_place.iterrows():
    try:
        df_tweet_place.loc[index,"location_type"] = df_tweet_place["bounding_box"][index].type
        df_tweet_place.loc[index,"long_1"] = df_tweet_place["bounding_box"][index].coordinates[0][0][0]
        df_tweet_place.loc[index,"lat_1"] = df_tweet_place["bounding_box"][index].coordinates[0][0][1]
        df_tweet_place.loc[index,"long_2"] = df_tweet_place["bounding_box"][index].coordinates[0][1][0]
        df_tweet_place.loc[index,"lat_2"] = df_tweet_place["bounding_box"][index].coordinates[0][1][1]
        df_tweet_place.loc[index,"long_3"] = df_tweet_place["bounding_box"][index].coordinates[0][2][0]
        df_tweet_place.loc[index,"lat_3"] = df_tweet_place["bounding_box"][index].coordinates[0][2][1]
        df_tweet_place.loc[index,"long_4"] = df_tweet_place["bounding_box"][index].coordinates[0][3][0]
        df_tweet_place.loc[index,"lat_4"] = df_tweet_place["bounding_box"][index].coordinates[0][3][1]
    except:
        print(f"Missing place object {index}")
        
# separate lats and longs in tweet_user
df_tweet_user["location_type"] = 0
df_tweet_user["Longitude"] = 0
df_tweet_user["Latitude"] = 0
for index, row in df_tweet_user.iterrows():
    try:
        df_tweet_user.loc[index,"location_type"] = df_tweet_user["location_type_and_coordinates"][index]["type"]
        df_tweet_user.loc[index,"Longitude"] = df_tweet_user["location_type_and_coordinates"][index]["coordinates"][0]
        df_tweet_user.loc[index,"Latitude"] = df_tweet_user["location_type_and_coordinates"][index]["coordinates"][1]
    except TypeError:
        print(f"Missing location object {index}")

# drop parsed columns
df_tweet_user.drop("location_type_and_coordinates",axis=1,inplace=True)
df_tweet_place.drop("bounding_box",axis=1,inplace=True)
```
